Файл moscow_places.csv:
* name — название заведения;

* lat — широта географической точки, в которой находится заведение;

* avg_bill — строка, которая хранит среднюю стоимость заказа в виде диапазона, например:
    * «Средний счёт: 1000–1500 ₽»;

    * «Цена бокала пива: 400–600 ₽». и так далее;

     * Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
     * Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
     * Если значения нет или оно не начинается с подстроки «Средний счёт», то в столбец ничего не войдёт.
* middle_coffee_cup — число с оценкой одной чашки капучино, которое указано только для значений из столбца avg_bill, начинающихся с подстроки «Цена одной чашки капучино»:
    * Если в строке указан ценовой диапазон из двух значений, в столбец войдёт медиана этих двух значений.
    * Если в строке указано одно число — цена без диапазона, то в столбец войдёт это число.
    * Если значения нет или оно не начинается с подстроки «Цена одной чашки капучино», то в столбец ничего не войдёт.
* chain — число, выраженное 0 или 1, которое показывает, является ли заведение сетевым (для маленьких сетей могут встречаться ошибки):
     * 0 — заведение не является сетевым
     * 1 — заведение является сетевым
* district — административный район, в котором находится заведение, например Центральный административный округ;
* seats — количество посадочных мест.

### Отобразите все заведения датасета на карте с помощью кластеров средствами библиотеки folium.

```python

# создаём пустой кластер, добавляем его на карту
marker_cluster = MarkerCluster().add_to(m)

# пишем функцию, которая принимает строку датафрейма,
# создаёт маркер в текущей точке и добавляет его в кластер marker_cluster
def create_clusters(row):
    Marker(
        [row['lat'], row['lng']],
        popup=f"{row['name']} {row['rating']}",
    ).add_to(marker_cluster)

# применяем функцию create_clusters() к каждой строке датафрейма
df.apply(create_clusters, axis=1)

# выводим карту
m
```

**На первом этапе** проекта мы ознакомились с данными, затем совершили **предобработку данных**: удалили дубликаты; решили оставить пропуски, чтобы не потерять важную информацию; создали 2 вспомогательных столбца: с название улицы и режимом работы заведениями.

Далее мы приступили к **анализу данных**.
Узнали какие заведения представлены в Москве, их особенности, в т.ч. вместимость, режим работы, рейтинг, сердний чек и другие. Построили визуализации различных зависимостей и на основании наших наблюдений пришли к следующим вывода:
** В Москве широко представлены все виды заведений общественного питания. Самые лакомые места - это центральный округ и северо-западный округ. Там обитают богатые местные, туристы, собираются друзья, что гарантирует большой средний чек и высокую проходимость. Пустовать заведения не будет. Большая конкуренция вынуждает оказывать первоклассный сервис, создавая зону притяжения для гостей. В таких районах популярны дорогие рестораны, кофейни, бары.
* Москвичи все чаще заказывают доставку еды на дом. Благодаря этому сети с удобными приложениями доставки, такие как доминос пицца и Додо, Яндекс лавка получают большое преимущество и покрывают большую долю рынка.
* Среди кофеен, пиццерий и булочных сети распространены больше всего. Их технологичность, централизованность дают им преимущества в этих сегментах. Частным предпринимателям лучше открывать бары, кафе и рестораны. Люди любят такие места за своеобразие, индивидуальность и оплачивают лояльностью.
* При открытие столовой нужно учитывать район. Желательно поближе к офисам, например в центре, и завлекать людей низкими ценами. Помещение должно вмещать в среднем 75 человек.
* Бары и пабы находят себя в проходимых улицах. Например, Ленинградский проспект, Ленинский проспект и Пятницкая улица. Люди очень лояльны к барам и готовы много тратить там. Чем дороже и интереснее бар, тем он привликательнее. В пятницу вечером в бары идут много людей и нужно позаботится, чтобы все желающие могли усесться, обычно это до 150 человек.
* Среди остальных заведений кафе самые популярные. Очень важно, чтобы до него можно было легко и просто добраться, например автомобилистам. Кафе прекрасно работают 24/7. Конкуренция хоть и большая, но часть заведений получает низкие отзывы. Москвичи готовы платить за качественный сервис, так что главное не поскупиться на качественном обслуживании клиентов. Ко всему прочему, не обязательно искать большое помещение. Люди в кафе обычно не задерживаются и приезжают в разное время.
* Сложнее всего открыть успешную частную булочную. Возможно проще всего работать по франшизе. Сети за счет своей централизованности могу экономить на издержках и производить продукцию в одной пекарне.
* К фастфудам люди относятся непритязательно. Они не ждут чего-то большего, кроме как быстро и недорого поесть. Подойдут места в спальных районах, где мало таких заведений.
* Рестораны популярны во всех районах. Особенно в центре. У таких заведений самый большой средний чек и сети не имеют преимуществ перед частниками. Москвичи готовы переплачивать за сервис и ставить таким ресторанам высокие оценки. Новый ресторан будет готов принмать одновременно 90 человек.

На **четвертом этапе** мы детализировали исследование, а именно по заказу инвесторов из «Shut Up and Take My Money» мы рассмотрели возможность открытия новой кофейни в Москве.
Мы ответили на вопросы:
- Сколько всего кофеен в датасете? В каких районах их больше всего, каковы особенности их расположения?
- Есть ли круглосуточные кофейни?
- Какие у кофеен рейтинги? Как они распределяются по районам?
- На какую стоимость чашки капучино стоит ориентироваться при открытии и почему?
- Какие рейтинги получают сетевые кофейни и обычные?
- В каких районах представлены сетевые кофейни?

И пришли к следующим рекомендациям:
- Всего кофеен в Москве 1 413. Это большой сегмент рынка общественного питания, третий по величине.
- Мы узнали, что больше всего кофеен в Центральном районе. Это самый богатый и активный район Москвы. Большая конкуренция и проходимость, высокие цены за чашку кофе и высокок качетсво обслуживания, иначе заведение провалится. Место для амбициозного проекта точно найдется.
- Можно выделить 4 региона-лидера по рейтингу - это Северный, Северо-западный, Юго-восточный и Юго-западный. Тут самые лояльные и неизбалованные клиенты. В этих округах не так много конкуренции, достаточная низкая цена чашки капучино и высокие рейтинги. Прекрасное место для открытия первой точки!
- Чашка капучино должна стоит от 150 до 200 рублей. Это не слишком дорого и привличет потенциальный клиентов, москвичи не очень любят переплачивать за кофе и высоко ценят недорогую цену за чашечку бодрости. Дешевле делать не имеет смысла, т.к. на отзывы это не повлияет положительным образом. Низкая цена может только насторожить.
- Круглосуточную кофейню открывать нет смысла. Их очень мало, популярностью они не пользуются. Это лишь создаст дополнительные риски. Такая идея будет хороша только около вокзалов и подобных мест, где много людей ищут спокойное и уютное место ночью.
- У несетевых кофеен рейтинг выше на 0,2 балла и нередко достигает высшей оценки в отличии от сетей. Это весомая причина открывать кофейню самостоятельно.
- Почти по всем округам Москвы распределение сетевых и несетевых кофеен 50 на 50. Стоит отметить юго-восточный округ. Здесь подавляющее большинство несетевых кофеен, а значит пока больше кислорода. Сетевые кофейни вытесняют обычные за счет своих ресурсов. Избегать стоит Западный округ, где сетевых кофеен уже 62%.
- Помещение под новую кофейню должно вмещать минимум 60 посетителей одноврменно. Это норма для несетевых кофеен. В проходимом месте желательно иметь 120 мест.

**Хорошим решением будет открыть кофейню в Юго-восточном округе со средней ценой чашки капучино 150-200 рублей. Успешная кофейня должна вмещать минимум 60 посетителей.**
